```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def plot_spectral_radius(txt_file) : 
    radius = []
    with open(txt_file, 'r') as myfile :
        # contents = myfile.readlines()
        # print(contents)
        for myline in myfile :
            l = myline.split()
            radius.append(float(l[0]))
    myfile.close()

    x = np.linspace(0, len(radius)-1, int(len(radius)/300), dtype = 'int')
    l = [radius[i] for i in x]
    plt.figure(figsize=[6,5])
    plt.plot(x, l, 'k', lw = 3, label = "spectral radius")
    # plt.yscale('log')
    plt.xlabel('Validation Iteration', fontsize = 20)
    plt.ylabel('Spectral Radius', fontsize = 20)
    plt.xticks(ticks = [0, 10000, 20000, 30000, 40000, 50000], labels=['0', '10k', '20k', '30k', '40k', '50k'], fontsize = 20)
    plt.yticks(fontsize = 20)
    plt.legend()
    plt.tight_layout()
    plt.savefig("spectral_radius", dpi = 400, transparent = True)
    # plt.title("Evolution of spectral radius through val epochs")
    plt.show()

plot_spectral_radius("spectral_radius_2.csv")
```
