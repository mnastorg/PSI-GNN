```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nb_nodes = [192.295, 486.555, 1825.255, 7099.205, 11083.9]

dss_val = [0.01447873539327702, 0.031776790220319526, 10.859008326260373, 200.14036320328714, 531.5822129678726]
# dsgps_val = [0.03654324361879844, 0.21842540359823034, 4.003945963401347, 149.3108893698454, 442.41210684478284]
psignn_val = [0.006040223998425063, 0.007614337908744346, 0.70235097792116, 67.69307005368174, 177.81265030145644]

plt.figure(figsize = [10,10])
plt.plot(nb_nodes, dss_val, color='black', marker = 'o', linestyle = 'dotted', markersize = 15, linewidth = 3, label = "DSS")
# plt.plot(nb_nodes, dsgps_val, color='black', marker = 'P', linestyle = 'dashed', markersize = 15, linewidth = 3, label = "DSGPS")
plt.plot(nb_nodes, psignn_val, color='black', marker = 's', linestyle = 'solid', markersize = 15, linewidth = 3, label = "$\Psi$-GNN")
plt.legend(fontsize=20)
plt.xlabel("Number of nodes", fontsize = 25)
plt.ylabel("MSE w/LU", fontsize = 25)
plt.xticks(np.linspace(0, 10000, 5, dtype = 'int'), fontsize = 25)
plt.yticks(fontsize = 25)
# plt.yscale("log")
plt.tight_layout()
plt.savefig("growing_geometry_comparative_dss_vs_psignn.png", dpi = 300, transparent = False)
```

```python
nb_nodes = [192.295, 486.555, 1825.255, 7099.205, 11083.9]
dss_rel = np.asarray([0.0138223, 0.021637723, 0.3060784, 0.7342763, 0.82398957])
dsgps_rel = np.asarray([0.02375119, 0.047807068, 0.2033599, 0.61351043, 0.72465074])
psignn_rel = np.asarray([0.01138645, 0.013293642, 0.05354459, 0.38044235, 0.418134])

plt.figure(figsize = [10,10])
plt.plot(nb_nodes, 100*dss_rel, color='black', marker = 'o', linestyle = 'dotted', markersize = 15, linewidth = 3, label = "DSS")
plt.plot(nb_nodes, 100*dsgps_rel, color='black', marker = 'P', linestyle = 'dashed', markersize = 15, linewidth = 3, label = "DSGPS")
plt.plot(nb_nodes, 100*psignn_rel, color='black', marker = 's', linestyle = 'solid', markersize = 15, linewidth = 3, label = "$\Psi$-GNN")
plt.legend(fontsize=20)
plt.xlabel("Number of nodes", fontsize = 25)
plt.ylabel("MSE w/LU", fontsize = 25)
plt.xticks(np.linspace(0, 11000, 5, dtype = 'int'), fontsize = 25)
plt.yticks(fontsize = 25)
plt.yscale("log")
plt.tight_layout()
plt.savefig("growing_relative.png", dpi = 500, transparent = True)
```
